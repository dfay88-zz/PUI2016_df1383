```python
# Import necessary packages
from __future__ import division, print_function
import pandas as pd
import pylab as pl
import os
%pylab inline

# Get NYC Open Data file path
DFDATA = os.getenv("DFDATA")
print (DFDATA)
```

```python
# Load csv file to pandas dataframe and print first 5 rows
df = pd.read_csv(DFDATA + "/7z8d-msnt/1414245990/7z8d-msnt")
df.head(5)
```

```python
# Remove the district column and print first 5 rows
df.drop(['District'],inplace=True, axis=1)
df.head(5)
```

```python
# Plot Enrollment Avg versus Percent Attendance Avg
df['YTD % Attendance (Avg)'] = df['YTD % Attendance (Avg)'].str.replace('%', '')
plt.scatter(df['YTD Enrollment(Avg)'],df['YTD % Attendance (Avg)'])
plt.xlabel('Average Number of Enrolled Students')
plt.ylabel('Average Attendance Percentage')
plt.title('Average Number of Students vs Average Attendance Percentage',y=1.05)
```

```python
# Remove outlier from dataframe and Plot
df = df[df['YTD Enrollment(Avg)'] <= 100000]
plt.scatter(df['YTD Enrollment(Avg)'],df['YTD % Attendance (Avg)'])
plt.xlabel('Average Number of Enrolled Students')
plt.ylabel('Average Attendance Percentage')
plt.title('Average Number of Students vs Average Attendance Percentage in NYC',y=1.05)
```

#### The above plot shows the relationship between the average number of enrolled students and the average attendance percentage.
#### Data obtained from CUSP Data Hub - School Attendance and Enrollment Statistics by District (2010-11)
