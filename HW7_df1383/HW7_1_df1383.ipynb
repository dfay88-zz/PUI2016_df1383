```python
# Import necessary packages

import pandas as pd
```

```python
# Read drug use dataset into pandas dataframe

df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/drug-use-by-age/drug-use-by-age.csv')
df.head()
```

```python
# Print columns

df.columns
```

```python
# Select relevant columns needed for plot

drug_use = df[['age', 'alcohol-use', 'alcohol-frequency', 'marijuana-use',
       'marijuana-frequency', 'cocaine-use', 'cocaine-frequency']]
drug_use = drug_use[drug_use.age !='65+']
```

```python
# Convert age column values that are ranges into average of range

drug_use1 = drug_use.replace('22-23', 22)
drug_use1 = drug_use1.replace('24-25', 24)
drug_use1 = drug_use1.replace('26-29', 26)
drug_use1 = drug_use1.replace('30-34', 30)
drug_use1 = drug_use1.replace('35-49', 35)
drug_use1 = drug_use1.replace('50-64', 50)
drug_use1 = drug_use1.replace('65+', 65)

drug_use1
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Scatter(
    x=drug_use1['age'],
    y=drug_use1['alcohol-frequency'],
    name='Frequency of Alcohol Use',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 3)
)
trace2 = go.Bar(
    x=drug_use1['age'],
    y=drug_use1['alcohol-use'],
    name='% who us Alcohol',
    yaxis='y2',
    marker = dict(
        color = ('rgb(205, 12, 24)'),
    ),
)

trace3 = go.Scatter(
    x=drug_use1['age'],
    y=drug_use1['marijuana-frequency'],
    name='Frequency of Marijuana Use',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        width = 3)
)
trace4 = go.Bar(
    x=drug_use1['age'],
    y=drug_use1['marijuana-use'],
    name='% who use Marijuana',
    yaxis='y2',
    marker = dict(
        color = ('rgb(22, 96, 167)'),
    ),
)

trace5 = go.Scatter(
    x=drug_use1['age'],
    y=drug_use1['cocaine-frequency'],
    name='Frequency of Cocaine Use',
    line = dict(
        color = ('rgb(12, 200, 24)'),
        width = 3)
)
trace6 = go.Bar(
    x=drug_use1['age'],
    y=drug_use1['cocaine-use'],
    name='% who use Cocaine',
    yaxis='y2',
    marker=dict(
        color = 'rgb(12, 200, 24)'),
)

data = [trace3, trace4, trace5, trace6]
layout = go.Layout(
    barmode ='group',
    title='Drug Use',
    yaxis=dict(
        showgrid=False,
        title='% of Individuals in Age Group (Bar)'
    ),
    xaxis=dict(
        showgrid=False,
        title='Age'
    ),
    yaxis2=dict(
        showgrid=False,
        title='Frequency of Use (Line)',
        overlaying='y',
        side='right'
    ),
    showlegend=True,
    legend=dict(
        x=100,
        y=1.225,
    ),
)


fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Drug Use')
```
