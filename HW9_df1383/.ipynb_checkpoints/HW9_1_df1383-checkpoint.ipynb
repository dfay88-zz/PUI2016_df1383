### Task 1:
Event detection: Identify the most prominent event. There is a very significant drop (>3-sigma) in all time series. Identify it, figure out the date (you know when the data starts and what the cadence is) and figure out what it is due to.

```python
# Sum over all stations and card types
avg_tot_ridership = np.nanmean(mta_fare, axis=(0,1))
print (avg_tot_ridership.shape)
```

```python
# Plot total average ridership
plt.figure(figsize=(15,8))
plt.plot(avg_tot_ridership)
plt.xlabel('Number of Weeks from 05/21/2010', fontsize=15)
plt.ylabel('Total Average Ridership', fontsize=15)
plt.title('Average Ridership from 05/21/2010 to 02/21/2014', fontsize=18)
```

#### Figure 1:
Figure 1 illustrates the total average ridership across all stations and card types in NYC for each week starting from 05/21/2010. Based on visual analysis, I would argue there is one distinct outlier in ridership around week 125 or so. Every other week ranges between 2000 - 3500 ridership while ridership falls to 500 during one week, which would indicate there may be an event which affected the subway during that time period.

```python
# Define threshold range as 3 sigma from mean
np.nanmean(avg_tot_ridership)
threshold = [np.nanmean(avg_tot_ridership) + 3 * np.nanstd(avg_tot_ridership), 
              np.nanmean(avg_tot_ridership) - 3 * np.nanstd(avg_tot_ridership)]
threshold
```

```python
# Plot all 23 card types total ridership
plt.figure(figsize=(15,8))
plt.plot(avg_tot_ridership, label = 'Avg Ridership')
plt.xlabel('Number of Weeks from 05/21/2010', fontsize=15)
plt.ylabel('Total Average Ridership', fontsize=15)
plt.title('Average Ridership from 05/21/2010 to 02/21/2014', fontsize=18)
plt.plot([0, 194], [threshold[1], threshold[1]], label = '<3 Sigma Threshold')
plt.plot([0, 194], [threshold[0], threshold[0]], label = '>3 Sigma Threshold')
plt.legend(loc=4)
```

#### Figure 2:
Figure 2 illustrates the total average ridership across all stations and card types in NYC for each week starting from 05/21/2010 including lines which indicate ridership values that are 3 standard deviations from the mean. These threshold lines help visualize weeks where ridership is more or less than three standard deviations from the mean and help us identify potential events which caused the significant change in ridership.

```python
# Find Average Ridership and Week where ridership is < Sigma from Mean
event_avg_ridership = avg_tot_ridership[np.where(avg_tot_ridership < threshold[1])]
event_week = (np.where(avg_tot_ridership < threshold[1])[0]).astype('int')
print ('Event Average Ridership: ', event_avg_ridership[0])
print ('Event Week: ', event_week[0]+1)
```

```python
# Convert Week Number to Date
week_range = pd.date_range('05/28/2010', periods=len(avg_tot_ridership), freq='w-fri')
event_date_start = week_range[event_week[0]]
event_date_end = week_range[event_week[0]+1]
print ('Event Start Date: ', event_date_start)
print ('Event End Date: ', event_date_end)
```

### Task 2:
Some of the time series are stationary, some have a complex structure, some show a downward trend: Identify the ridership types that have steadily increased in popularity, and that have steadily decreased. by how much? (e.g what is the ratio of usage in the first 10 and last 10 weeks)

```python
# Calculate percent of ridership for each card type and load into pandas dataframe
sum_card_ridership = pd.DataFrame(np.nansum(mta_fare, axis=0)).T
tot_sum_card_ridership = sum_card_ridership.sum(axis=1)
for i in range(len(sum_card_ridership.T)):
    for j in range(len(sum_card_ridership)):
        sum_card_ridership[i][j] = sum_card_ridership[i][j]/tot_sum_card_ridership[j]
sum_card_ridership.head()
```

```python
# Fit line to each ridership type and plot card types that show increasing and decreasing trends
ridetype = np.array(['10t', '14d', '1d', '2t', '30d', '7d', 'ada', 'afas', 'exp','ez',
            'ff','mc','mr','month','rfm','rr','sen','spec','tcmc','t',
            'umlim','xbus','stud'])

lr = linear_model.LinearRegression()
rm_sum_card_ridership = pd.rolling_mean(sum_card_ridership, 10)

for column in rm_sum_card_ridership:
    df = pd.DataFrame(rm_sum_card_ridership[column]).reset_index()
    df.columns = ['week', 'rides']
    res = smf.ols(formula='rides ~ week', data=df).fit()
    avg_beg = np.nanmean(sum_card_ridership[column][:10])
    avg_end = np.nanmean(sum_card_ridership[column][:-10])
    ridership_ratio = avg_end / avg_beg
    if (res.rsquared_adj > .6) & (ridership_ratio > 1.05 or ridership_ratio < .95):
        plt.figure(figsize = (12,8))
        plt.plot(res.predict(), label='fitted line')
        plt.title("Card Type: %s, Adj R-Squared: %s, Ridership Ratio: %s"%(ridetype[column],
                                                                           res.rsquared_adj, ridership_ratio), fontsize=16)
        plt.xlabel("Week Number", fontsize=15)
        plt.ylabel("Ridership", fontsize=15)
        plt.plot(rm_sum_card_ridership[column], label="original")
        plt.legend(loc=2)
```

### Trend Analysis:
The above plots illustrate all the card types that were determined to have a strong positive or negative trend in regards to ridership over the 194 weeks. First a ratio was calculated by taking the average percent ridership for the first 10 and last 10 weeks and dividing them. A ridership ratio of greater than 1 indicates a positive trend and a ridership ratio less than 1 indicates a negative trend. Next card types that had a ridership ratio greater than 1.05 and less than .95 were selected for further analysis. Then the selected card types were smoothed using rolling mean and had a line fitted using linear regressoin. Finally, card types where the adjusted r-squared was greater than .6 were deemed to have a strong positive or negative trend.

#### Card types which have steadily increased:
* 14d
* 7d
* rr
* sen
* tcmc
* xbus

#### Card types which have steadily decreased:
* 2t
* ada
* month

### Task 3:
Several stations show a prominent annual periodicity. Identify the 4 stations (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)

```python
# Calculate mean of card type ridership for each station
avg_station_ridership = pd.DataFrame(np.nanmean(mta_fare, axis=1)).T
sum_station_ridership = pd.DataFrame(np.nansum(mta_fare, axis=1)).T
avg_station_ridership.head()
```

```python
# Perform fourier transformation on all stations and create new array with 5th element of each fourier tranformation
fourier_list = np.array([])
for column in sum_station_ridership:
    fourier_trans = np.abs(np.fft.rfft(sum_station_ridership[column]))
    fourier_list = np.append(fourier_list, fourier_trans[4])
    
fourier_list = pd.DataFrame(fourier_list)
fourier_list.sort_values(0, ascending=False, inplace=True)
#fourier_list.dropna(inplace=True)
fourier_list.head()
```

```python
# Plot power spectrum for the 4 stations with the highest power value for the 5th element of the fourier transformation
for column in fourier_list[:4].T:
    N = len(avg_station_ridership)
    fourier = np.abs(np.fft.rfft(sum_station_ridership[column]))
    fig = plt.figure(figsize=(15,6))
    plt.plot(week_range, sum_station_ridership[column])
    plt.suptitle('Ridership Over Time', y=1.005, fontsize=18)
    plt.title('Station: %s'%column, fontsize=15)
    plt.xlabel("Date", fontsize=18)
    plt.ylabel("Ridership", fontsize=18)
    
    fig = plt.figure(figsize=(15,6))
    ax=fig.add_subplot(111)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (fourier)[1:], 'o', ms=8)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (fourier)[1:], '-')
    ax.set_xlim(0,2.0/12)
    ax.set_xticklabels([ "%s"%(1/4/f) for f in ax.get_xticks()], fontsize=15)
    plt.suptitle('Ridership Power Spectrum', y=1.005, fontsize=18)
    plt.title('Station: %s'%column, fontsize=15)
    plt.xlabel("Period (months)", fontsize=18)
    plt.ylabel("Power", fontsize=18)
```

### Periodicity Analysis:
The above plots illustrate ridership over the 194 weeks and the corresponding power spectrum plot for the four stations with the highest annual periodicity. The stations were determined by using the fourier transformation and comparing the values corresponding to the 5th element.
