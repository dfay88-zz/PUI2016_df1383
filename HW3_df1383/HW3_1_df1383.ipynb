#### Import Necessary Packages

```python
%reset
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import matplotlib.mlab as mlab
from scipy.stats import norm
%matplotlib inline
```

### Distribution 1: Chisquared

```python
# Create 100 different sample sizes
sample_size = np.arange(10,2000,20)
sample_size
```

```python
# Create 100 Chisquared distributions of different sample sizes, calculate mean and plot sample size by sample mean
mean = 100
dict_dist = {}
dict_dist['Chisq'] = {}

for n in sample_size:
    dict_dist['Chisq'][n] = np.random.chisquare(mean, size = n).mean()
    
chisq_plot = plt.figure(figsize=(10,6)).add_subplot(111)

for n in dict_dist['Chisq']:
    chisq_plot.plot(n, dict_dist['Chisq'][n], 'o')
    chisq_plot.set_xlabel('sample size', fontsize=18)
    chisq_plot.set_ylabel('sample mean', fontsize=18)
    chisq_plot.set_title('Chisquared', fontsize=18)
    chisq_plot.plot([min(sample_size), max(sample_size)], [mean, mean], 'k')
```

The above plot helps prove part of the central limit theorem by illustrating that as the size of the sample increases the sample mean converges to the population mean.

```python
# Plot histogram of sample mean by size of the sample
plt.figure(figsize=(10, 6))
chisq_means = []
for n in dict_dist['Chisq']:
    chisq_means.append(dict_dist['Chisq'][n])

mu,sigma = norm.fit(chisq_means)
n,bins,patches = plt.hist(chisq_means,bins=30,normed=True)
y = mlab.normpdf(bins, mu, sigma)

plt.title('Chisquared Sample Mean Distribution, N=100', fontsize = 18)
plt.xlabel('Sample Mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
plt.plot(bins, y, 'r--', linewidth=2)
```

The above plot helps prove part of the central limit theorem by illustrating that the means of a sufficiently large enough number of samples will be normally distributed.
